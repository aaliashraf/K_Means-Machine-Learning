```python
import random
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
print("done")
```

```python
#getting dataset
customers_df = pd.read_csv('customer_segmentation.csv')
customers_df.head()
```

```python
#removal of the attribute as euclidean distance is not directly applicable to this datatype
df = customers_df.drop('Address', axis=1)
df.head()
```

```python
#for normalization see more details on StandardScaler documentation
from sklearn.preprocessing import StandardScaler

X = df.values[:,1:]
X = np.nan_to_num(X)
cluster_dataset = StandardScaler().fit_transform(X)
cluster_dataset
```

```python
#here the sklearn k_means applies
num_clusters = 3

k_means = KMeans(init="k-means++", n_clusters=num_clusters, n_init=12)
k_means.fit(cluster_dataset)
labels = k_means.labels_

print(labels)
```

```python
#Customers belonging to which cluster 
df["Clusters"] = labels
df.head(5)
```

```python
#taking the mean of cluster attributs
df.groupby('Clusters').mean()
```
